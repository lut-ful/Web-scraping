### Here I collected the data from list(for demo purpose collected only 10 row) of similar website using simple method. Then exported it in CSV file.

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_clinic_name(clinic_id):
    url=f'https://{clinic_id}.portal.athenahealth.com/'
    response=requests.get(url)
    html=response.text
    soup=BeautifulSoup(html,'html.parser')
    clinic_name=soup.find_all('h1')[-1].text.strip()
    return clinic_name
```

```python
start=12690
end=12700
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
df.columns
```

```python
import numpy as np
```

```python
master_list=[]
for clinic_id in range(start,end+1):
    name=get_clinic_name(clinic_id)
    if name!="Sorry, we can't find that practice. Make sure you typed the right address." and name!="Payment Confirmation":
        data_dict={'Clinic Id':clinic_id,'Clinic Name':name}
        master_list.append(data_dict)
    
```

```python
master_list
```

```python
df=pd.DataFrame(master_list)
df
```

```python
df.to_csv('finds_the_website_same_format.csv',index=False)
```
